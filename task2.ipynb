```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
train_data = pd.read_csv('train.csv')

train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data.drop(columns=['Cabin'], inplace=True)

print("Descriptive Statistics:")
print(train_data.describe())

print("\nMissing Values:")
print(train_data.isnull().sum())
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.barplot(data=train_data, x='Pclass', y='Survived', palette='Blues')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(data=train_data, x='Sex', y='Survived', palette='Pastel1')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(train_data['Age'], kde=True, bins=30, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=train_data, x='Survived', y='Age', palette='Set2')
plt.title('Survival by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=train_data, x='Embarked', y='Survived', palette='Set3')
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(train_data[train_data['Survived'] == 1]['Fare'], bins=30, color='blue', label='Survived', kde=True)
sns.histplot(train_data[train_data['Survived'] == 0]['Fare'], bins=30, color='red', label='Did not Survive', kde=True)
plt.title('Survival by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(data=train_data, x='Embarked', y='Survived', palette='Set3')
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(train_data[train_data['Survived'] == 1]['Fare'], bins=30, color='blue', label='Survived', kde=True)
sns.histplot(train_data[train_data['Survived'] == 0]['Fare'], bins=30, color='red', label='Did not Survive', kde=True)
plt.title('Survival by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()
```
